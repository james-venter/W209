```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("players_20.csv")
```

```python
df.head()
```

```python
count = df.groupby('nationality')['sofifa_id'].count().reset_index()
age = df.groupby('nationality')['age'].mean().round(1).reset_index()

summary = count.merge(age)
summary.columns =['Nationality', 'Count', 'Avg Age']
summary.head(10)
```
